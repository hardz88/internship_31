# Q-2 Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
import pandas as pd   #to create dataframe
import requests       #to send the request to the URL
from bs4 import BeautifulSoup #to get the content in the form of HTML
import numpy as np  # to count the values (in our case)
```

```python
#assigning the URL with variable name url
url = 'https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc'
#request allow you to send HTTP request
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

#creating an empty list, so that we can append the values
movie_name = []
year = []
rating = []

#storing the meaningfull required data in the variable
movie_data = soup.findAll('div', attrs= {'class': 'lister-item mode-advanced'})

#calling one by one using for loop
for store in movie_data:
    name = store.h3.a.text
    movie_name.append(name)
    
    year_of_release = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '')
    year.append(year_of_release)
    
    rate = store.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n', '')
    rating.append(rate)
    
    #since, gross and votes have same attributes, that's why we had created a common variable and then used indexing
    value = store.find_all('span', attrs = {'name': 'nv'})
    
    
    #creating a dataframe using pandas library
movie_DF = pd.DataFrame({'Name of movie': movie_name, 'Year of relase': year, 'Movie Rating': rating,})
```

# Q3 Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame

```python
#assigning the URL with variable name url
url = 'https://www.imdb.com/india/top-rated-indian-movies/'
#request allow you to send HTTP request
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

#creating an empty list, so that we can append the values
movie_name = []
year = []
rating = []
#storing the meaningfull required data in the variable
#movie_data = soup.findAll('tbody', attrs= {'class': 'lister-list'})

movie_data = soup.find('tbody', attrs= {'class': 'lister-list'}).findAll('tr')
movie_data

for store in movie_data:
    titledata = store.find('td', class_ = 'titleColumn');
    name = titledata.a.text.replace('(', '').replace(')', '')
    movie_name.append(name)
    
    years = titledata.span.text.replace('(', '').replace(')', '')
    year.append(years)
   
    ratings = store.find('td', class_ = 'ratingColumn imdbRating').strong.text.replace(')', '')
    rating.append(ratings)
    
movie_DF = pd.DataFrame({'Name of movie': movie_name, 'Year of relase': year, 'Movie Rating': rating})
```

# Q-4 Write s python program to display list of respected former presidents of India(i.e. Name , Term of office)

```python
#assigning the URL with variable name url
url = 'https://presidentofindia.nic.in/former-presidents.htm'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

president_names = []
term_office = []
former_data = soup.find('ul', attrs= {'class': 'listing cf'}).findAll('li')

for fdata in former_data:
    name = fdata.h3.text.split(' (')[0]
    president_names.append(name)
    terms = fdata.p.text.replace('Term of Office: ', '')
    term_office.append(terms)
    
president_df = pd.DataFrame({'President Names': president_names, 'Term of office': term_office})
president_df.head()
```

# Q-5 Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
b) Top 10 ODI Batsmen along with the records of their team and rating.
c) Top 10 ODI bowlers along with the records of their team and rating.

```python
#  a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

team_names = []
odi_matches = []
match_points = []
runs_rating = []

top1 = soup.find('tr', attrs = {'class': 'rankings-block__banner'})

td_elems = top1.findAll('td')
team_names.append(top1.find_next('span').find_next('span').text)
odi_matches.append(td_elems[2].text)
match_points.append(td_elems[3].text)
runs_rating.append(td_elems[4].text.replace('\n', ''))

top_mens = soup.findAll('tr', attrs = {'class': 'table-body'})
for mens_odi in top_mens:
    tname = mens_odi.find("span", {"class":"u-hide-phablet"}).text
    team_names.append(tname)
    name = mens_odi.findAll('td', attrs = {'class': 'u-center-text'})
    odi_matches.append(name[0].text)
    match_points.append(name[1].text)    
    ratings = mens_odi.find('td', attrs = {'class': 'rating'}).text
    runs_rating.append(ratings)
    
mensdata = pd.DataFrame({'Team': team_names, 'Matches': odi_matches, 'Points': match_points, 'Rating': runs_rating})
mensdata.head(10)
```

```python
#  b) Top 10 ODI Batsmen along with the records of their team and rating

url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

odi_matches = []
match_points = []
runs_rating = []
teams = []

batsmans = soup.find('div', attrs = {'data-title': 'ODI Batting Rankings'});

first_name = batsmans.find("div",class_="rankings-block__banner--name").get_text(strip=True,separator="")
odi_matches.append(first_name)

first_team = batsmans.find("div",class_="rankings-block__banner--nationality").get_text(strip=True,separator="")
teams.append(first_team)

first_rate = batsmans.find("div",class_="rankings-block__banner--rating").text
runs_rating.append(first_rate)

top_mens = batsmans.findAll('tr', attrs = {'class': 'table-body'})
for mens_odi in top_mens:
   # name = mens_odi.span.text
    player_name = mens_odi.find('td', attrs = {'class': 'table-body__cell name'}).text.replace('\n', '')
    odi_matches.append(player_name)   
    team_name = mens_odi.find('span', attrs = {'class': 'table-body__logo-text'})
    teams.append(team_name.text)
    ratings = mens_odi.find('td', attrs = {'class': 'rating'})
    runs_rating.append(ratings.text)
    
mensdata = pd.DataFrame({'Batsman': odi_matches, 'Team': teams, 'Rating': runs_rating})
mensdata.head(10)
```

```python
#  c) Top 10 ODI bowlers along with the records of their team and rating

url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

odi_matches = []
match_points = []
runs_rating = []
teams = []

batsmans = soup.find('div', attrs = {'data-title': 'ODI Bowling Rankings'});

first_name = batsmans.find("div",class_="rankings-block__banner--name").get_text(strip=True,separator="")
odi_matches.append(first_name)

first_team = batsmans.find("div",class_="rankings-block__banner--nationality").get_text(strip=True,separator="")
teams.append(first_team)

first_rate = batsmans.find("div",class_="rankings-block__banner--rating").text
runs_rating.append(first_rate)

top_mens = batsmans.findAll('tr', attrs = {'class': 'table-body'})

for mens_odi in top_mens:
   # name = mens_odi.span.text
    player_name = mens_odi.find('td', attrs = {'class': 'table-body__cell name'}).text.replace('\n', '')
    odi_matches.append(player_name)   
    team_name = mens_odi.find('span', attrs = {'class': 'table-body__logo-text'})
    teams.append(team_name.text)
    ratings = mens_odi.find('td', attrs = {'class': 'rating'})
    runs_rating.append(ratings.text)
    
mensdata = pd.DataFrame({'Bowler': odi_matches, 'Team': teams, 'Rating': runs_rating})
mensdata.head(10)
```

# Q-7 Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world
i) Headline ii) Time iii) News Link

```python
url = 'https://www.cnbc.com/world/?region=world'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
print(soup.prettify())

# Scrap Headlines

scraped_headlines = soup.find_all('a', class_='LatestNews-headline')
scraped_headlines

headlines = []

for hd in scraped_headlines:
    hd = hd.get_text().replace('\n','')
    headlines.append(hd)
headlines
```

```python
# Scrap Time

scraped_time = soup.find_all('time', class_='LatestNews-timestamp')
scraped_time

time = []

for t in scraped_time:
    t = t.get_text().replace('\n','')
    time.append(t)
time
```

```python
# Scrap Newslink

scraped_News_link = soup.find_all('a', class_="LatestNews-headline")
scraped_News_link

news_link = []

for nl in scraped_News_link:
    news_link.append(nl['href'])
news_link
```

# Q-9 Write a python program to scrape mentioned details from dineout.co.in:
i) Restaurant name
ii) Cuisine
iii) Location 
iv) Ratings
v) Image URL

```python
url = 'https://www.dineout.co.in/delhi-restaurants/buffet-special'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
# Scrap Restaurant Name
scraped_restaurant_name = soup.find_all('a', class_= 'restnt-name ellipsis')
scraped_restaurant_name

restaurant_names = []

# parse Restaurant Name
for rn in scraped_restaurant_name:
    rn = rn.get_text().replace('\n','')
    restaurant_names.append(rn)
restaurant_names
```

```python
# Scrap Cuisines
scraped_cuisines = soup.find_all('div', class_= 'filter-component-wrap cuisine-wrap')
scraped_cuisines

# Create empty list
cuisines = []

# parse Cuisines
for c in scraped_cuisines:
    c = c.get_text().replace('\n','')
    cuisines.append(c)
cuisines
```

```python
# Scrap Location
scraped_location = soup.find_all('div', class_= 'restnt-loc ellipsis')
scraped_location

locations = []

# parse location

for l in scraped_location:
    l = l.get_text().replace('\n','')
    locations.append(l)
locations
```

```python
# scrap Ratings
scraped_ratings = soup.find_all('div', class_= 'restnt-rating rating-4')
scraped_ratings

ratings = []

# parse location

for r in scraped_ratings:
    r = r.get_text().replace('\n','')
    ratings.append(r)
ratings
```

```python
# scrap Image URL
scraped_image_url = soup.find_all('img', class_= 'no-img')
scraped_image_url

image_url = []

#parse Image URL

for i in scraped_image_url:
    image_url.append(i['data-src'])
image_url
```

# Q-10 Write a python program to scrape the details of top publications from Google Scholar from 
https://scholar.google.com/citations?view_op=top_venues&hl=en
i) Rank 
ii) Publication
iii) h5-index
 iv) h5-median

```python
url = 'https://scholar.google.com/citations?view_op=top_venues&hl=en'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
#print(soup.prettify())

ranks = []
publications = []
indexes = []
medians = []

publication = soup.find('table', attrs = {'id': 'gsc_mvt_table'}).findAll('tr');
#publication

for pubs in publication[1:]:
   # name = mens_odi.span.text
   # print(pubs)
    mydata = pubs.findAll('td')
    ranks.append(mydata[0].text)
    publications.append(mydata[1].text)
    indexes.append(mydata[2].a.text)
    medians.append(mydata[3].span.text)
    
    
data = pd.DataFrame({'Ranks': ranks, 'Publication': publications, 'h5-index': indexes, 'h5-median': medians})
data.head(10)
```
